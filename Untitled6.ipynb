1. Bi-level and Multi-level Confusion Matrix


```python
from sklearn.metrics import confusion_matrix
import numpy as np

# Binary example
y_true = [0, 1, 0, 1, 1]
y_pred = [0, 1, 0, 0, 1]
print("Binary Confusion Matrix:\n", confusion_matrix(y_true, y_pred))

# Multi-class example
y_true = [0, 1, 2, 2, 0, 1]
y_pred = [0, 2, 1, 2, 0, 1]
print("\nMulti-class Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
```

2. Verification of Overfitting Issue

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = DecisionTreeClassifier(max_depth=None)  # Overfitting risk
model.fit(X_train, y_train)

print("Train Accuracy:", accuracy_score(y_train, model.predict(X_train)))
print("Test Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

X, y = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

model = LogisticRegression(max_iter=200).fit(X_train, y_train)
print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])  # y = 2x
m, b = 0, 0
lr = 0.01
epochs = 1000

for _ in range(epochs):
    y_pred = m * X + b
    dm = (-2/len(X)) * sum(X * (y - y_pred))
    db = (-2/len(X)) * sum(y - y_pred)
    m -= lr * dm
    b -= lr * db

print(f"Learned model: y = {m:.2f}x + {b:.2f}")
```
