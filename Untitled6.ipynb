5.Demonstrating the performance of gradient descent algorithm using Python.



```python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])  # y = 2x
m, b = 0, 0
lr = 0.01
epochs = 1000

for _ in range(epochs):
    y_pred = m * X + b
    dm = (-2/len(X)) * sum(X * (y - y_pred))
    db = (-2/len(X)) * sum(y - y_pred)
    m -= lr * dm
    b -= lr * db

print(f"Learned model: y = {m:.2f}x + {b:.2f}")
```

7.Verifying the performance of a image pre-processing with watershed algorithm

using Python.

```python
import cv2

import numpy as np

import matplotlib.pyplot as plt

from skimage import data

img = data.coins()

gray = img.astype(np.uint8)

kernel = np.ones((3,3), np.uint8)

opening = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel, iterations=2)

sure_bg = cv2.dilate(opening, kernel, iterations=3)

dist = cv2.distanceTransform(opening, cv2.DIST_L2, 5)

_, sure_fg = cv2.threshold(dist, 0.4 * dist.max(), 255, 0)

sure_fg = np.uint8(sure_fg)

unknown = cv2.subtract(sure_bg, sure_fg)

_, markers = cv2.connectedComponents(sure_fg)

markers = markers + 1

markers[unknown == 255] = 0

color_img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

markers = cv2.watershed(color_img, markers)

color_img[markers == -1] = [255, 0, 0]

plt.imshow(color_img)

plt.title("Watershed Segmentation")

plt.axis('off')

plt.show()
```

10: Performance evaluation of Linear regression Classifier using Breast cancer or

Iris database.

```python
import numpy as np
from sklearn.datasets import load_breast_cancer, load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

# Choose dataset: 'breast_cancer' or 'iris'
dataset_name = 'breast_cancer'

# Load dataset
if dataset_name == 'breast_cancer':
    data = load_breast_cancer()
    X, y = data.data, data.target  # Binary target: 0 or 1
elif dataset_name == 'iris':
    data = load_iris()
    X, y = data.data, data.target  # Multi-class: 0, 1, 2
else:
    raise ValueError("Dataset not recognized")

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict continuous output
y_pred_continuous = model.predict(X_test)

# Convert regression output to discrete class labels
# For binary classification, threshold at 0.5
if dataset_name == 'breast_cancer':
    y_pred = (y_pred_continuous >= 0.5).astype(int)
else:
    # For multi-class (e.g., Iris), round to nearest integer
    y_pred = np.rint(y_pred_continuous).astype(int)
    # Clip predictions to valid label range
    y_pred = np.clip(y_pred, y.min(), y.max())

# Evaluate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Dataset: {dataset_name}")
print(f"Accuracy: {accuracy:.4f}")
print("Confusion Matrix:")
print(cm)
```

16.Implementing the Hyper parameter Tuning on a chosen neural network scheme and verifying the change in the attained output.



```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import accuracy_score

# Load data
iris = load_iris()
X = iris.data
y = LabelBinarizer().fit_transform(iris.target)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Hyperparameter options
hidden_layers = [1, 2]
neurons_per_layer = [8, 16]
learning_rates = [0.01, 0.1]

# Simple grid search
best_acc = 0
best_params = {}

for n_layers in hidden_layers:
    for n_neurons in neurons_per_layer:
        for lr in learning_rates:
            # Build model
            model = Sequential()
            model.add(Dense(n_neurons, input_shape=(X.shape[1],), activation='relu'))
            for _ in range(n_layers - 1):
                model.add(Dense(n_neurons, activation='relu'))
            model.add(Dense(y.shape[1], activation='softmax'))

            model.compile(optimizer=Adam(learning_rate=lr), loss='categorical_crossentropy', metrics=['accuracy'])
            model.fit(X_train, y_train, epochs=20, verbose=0)

            # Evaluate
            y_pred = np.argmax(model.predict(X_test), axis=1)
            y_true = np.argmax(y_test, axis=1)
            acc = accuracy_score(y_true, y_pred)
            print(f"Layers: {n_layers}, Neurons: {n_neurons}, LR: {lr}, Accuracy: {acc:.2f}")

            # Save best
            if acc > best_acc:
                best_acc = acc
                best_params = {'layers': n_layers, 'neurons': n_neurons, 'lr': lr}

print("Best parameters:", best_params)
print("Best accuracy:", best_acc)
```

17.Implementing Convolutional Neural Network (CNN) models to execute the binary classification using Softmax classifier using Python



```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

# Simulated data: replace with your real images and labels
X = np.random.rand(100, 32, 32, 1)  # 100 images, 32x32 grayscale
y = np.random.randint(0, 2, 100)    # 0 or 1 labels
y = to_categorical(y, 2)            # converts to two-class softmax format

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Simple CNN model
model = Sequential([
    Conv2D(8, (3,3), activation='relu', input_shape=(32,32,1)),
        MaxPooling2D((2,2)),
            Flatten(),
                Dense(8, activation='relu'),
                    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5, batch_size=8, verbose=1)
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f'Test Accuracy: {acc:.2f}')
```

19.Implementing Convolutional Neural Network (CNN) models to achieve the

segmentation (UNet) using Python

```python
import tensorflow as tf
from tensorflow.keras import layers, Model

def simple_unet(input_shape=(128,128,1)):
    inputs = layers.Input(input_shape)

    # Encoder
    c1 = layers.Conv2D(32, 3, activation="relu", padding="same")(inputs)
    p1 = layers.MaxPooling2D()(c1)

    c2 = layers.Conv2D(64, 3, activation="relu", padding="same")(p1)

    # Decoder
    u1 = layers.UpSampling2D()(c2)
    m1 = layers.Concatenate()([u1, c1])
    c3 = layers.Conv2D(32, 3, activation="relu", padding="same")(m1)

    outputs = layers.Conv2D(1, 1, activation="sigmoid")(c3)

    return Model(inputs, outputs)

model = simple_unet()
model.compile(optimizer="adam", loss="binary_crossentropy")

# Dummy data
import numpy as np
x = np.random.rand(10,128,128,1)
y = np.random.rand(10,128,128,1)

model.fit(x, y, epochs=3)
```

24.Performance verification of NN-model using the spiral data.

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

def spiral(n_points, noise=0.2):
    n = np.sqrt(np.random.rand(n_points)) * 780 * (2*np.pi)/360
    d1x = -np.cos(n)*n + np.random.rand(n_points)*noise
    d1y = np.sin(n)*n + np.random.rand(n_points)*noise
    return np.vstack((d1x, d1y)).T

X = spiral(1000)
y = np.random.randint(0,2,1000)

model = Sequential([
    Dense(32, activation='relu', input_shape=(2,)),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss="binary_crossentropy", metrics=['accuracy'])
model.fit(X, y, epochs=5)
```
